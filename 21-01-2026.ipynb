### **Función y Propósito:**
El objetivo principal de este `ipynb` es transformar grandes volúmenes de registros contables en **insights estratégicos y accionables**. Su función es:

1.  **Auditar la Eficiencia del Gasto:** Identificar dónde y cómo se está utilizando el capital de la organización.
2.  **Detectar Tendencias y Patrones:** Analizar la evolución mensual, cuatrimestral y trimestral del gasto, así como la composición de costos por categoría, departamento y función de negocio.
3.  **Identificar Oportunidades de Optimización Presupuestaria:** Poner de manifiesto áreas de ineficiencia, riesgos de concentración (en proveedores, empleados, categorías) y posibles desviaciones respecto a una gestión financiera óptima.
4.  **Facilitar la Toma de Decisiones Estratégicas:** Proporcionar una base de datos visual y analítica para la planificación financiera de 2026, la gestión del *burn rate*, la optimización de la estructura de costos y la asignación de recursos.

```python
# @title Configuración de gráficos y Preprocesamiento
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go
import plotly.io as pio
from plotly.subplots import make_subplots

# --- 1. CONFIGURACIÓN DE DISEÑO Y MARCA (NOMII) ---
nomii_colors = {
    'primary': '#003366',   # Azul Profundo
    'secondary': '#20C6B6', # Turquesa Clínico
    'accent': '#FFCC00',    # Dorado
    'text': '#333333',      # Gris oscuro
    'background': '#F9F9F9',# Fondo suave
    'light_blue': '#4D7EA8',
    'pale_blue': '#B3D9EA'
}

nomii_palette = [nomii_colors['primary'], nomii_colors['secondary'], nomii_colors['light_blue'], nomii_colors['accent'], nomii_colors['pale_blue']]
sns.set_theme(style="whitegrid")
sns.set_palette(sns.color_palette(nomii_palette))
plt.rcParams['figure.figsize'] = (12, 6)
pio.templates["nomii_theme"] = go.layout.Template(
    layout=go.Layout(font={'color': nomii_colors['text']}, colorway=nomii_palette, paper_bgcolor=nomii_colors['background'], plot_bgcolor=nomii_colors['background'])
)
pio.templates.default = "nomii_theme"

# --- 2. PREPROCESAMIENTO DE DATOS (SOLO 2025) ---
pd.options.display.float_format = '{:,.2f} €'.format

# 1. Copia de seguridad
df = df.copy()
df['Date'] = pd.to_datetime(df['Date'])

# ==============================================================================
# ⚠️ CORRECCIÓN CRÍTICA DE MONEDA (Aquí arreglamos el error de los millones)
# ==============================================================================
# Calculamos la columna Amount in EUR manualmente usando la tasa de cambio.
# Esto ignora la columna vacía del Excel y arregla los montos en CLP.
df['Amount in EUR'] = df['Original Amount'] * df['ER to EUR']
# ==============================================================================

# 2. FILTRO DE FECHA: Nos quedamos solo con 2025
df = df[df['Date'].dt.year == 2025].copy()

# 3. Crear columnas temporales
df['YearMonth'] = df['Date'].dt.to_period('M')
df['YearMonth_Str'] = df['Date'].dt.strftime('%Y-%m')

# 4. Crear columna de Gasto Absoluto (Ahora sí usará los Euros correctos)
df['Gasto_Abs'] = df['Amount in EUR'].abs()

# 5. Ordenar por fecha
df = df.sort_values('Date')

print("✅ Corrección de divisas aplicada: CLP convertidos a EUR correctamente.")
print("✅ Filtro aplicado: Trabajando únicamente con datos del año 2025.")
print(f"✅ Registros restantes: {len(df)}")

# Validación rápida: Mostrar si queda algún monto gigante (Error si sale algo > 50k)
print("\n--- Verificación de Montos Máximos (No deberían superar 50.000€) ---")
display(df.sort_values(by='Gasto_Abs', ascending=False)[['Date', 'Counterparty', 'Original Amount', 'Amount in EUR']].head(3))
```

```python
# @title Evolución del 2025 con Línea de Variación %
import plotly.graph_objects as go
import plotly.express as px
import plotly.io as pio
import pandas as pd

# Configuración del tema global
pio.templates.default = "nomii_theme"

# --- 1. Preparación de datos ---
df_periodos = df.copy()
df_periodos['YearMonth_Str'] = df_periodos['YearMonth'].astype(str)

# Agrupamos solo por mes
df_macro = df_periodos.groupby('YearMonth_Str')['Gasto_Abs'].sum().reset_index()

# Calculamos variación y rellenamos el primer NaN
df_macro['Variacion_Pct'] = df_macro['Gasto_Abs'].pct_change() * 100
df_macro['Variacion_Pct'] = df_macro['Variacion_Pct'].fillna(0)

# Etiquetas para la línea (evitamos el 0% inicial visualmente)
etiquetas_variacion = [
    "" if i == 0 else f"{v:+.1f}%"
    for i, v in enumerate(df_macro['Variacion_Pct'])
]

fig_macro = go.Figure()

# --- 2. Construcción del Gráfico con Mejoras Visuales ---

# Barras: Gasto Total
fig_macro.add_trace(go.Bar(
    x=df_macro['YearMonth_Str'],
    y=df_macro['Gasto_Abs'],
    name='Gasto Total',
    text=df_macro['Gasto_Abs'],
    texttemplate='%{y:,.0f}€',
    textposition='auto',
    # MEJORA: Forzamos texto blanco y negrita para que se lea siempre sobre el azul
    textfont=dict(color='white', weight='bold')
))

# Línea: Variación Porcentual
fig_macro.add_trace(go.Scatter(
    x=df_macro['YearMonth_Str'],
    y=df_macro['Variacion_Pct'],
    name='Variación % Mensual',
    yaxis='y2',
    mode='lines+markers+text',
    text=etiquetas_variacion,
    textposition="top center",
    # Configuración Texto
    textfont=dict(color='#FFCC00', size=11),
    line=dict(width=2),
    # MEJORA: Cliponaxis evita que el texto se corte si se sale del marco
    cliponaxis=False
))

# --- 3. Layout ---
fig_macro.update_layout(
    title=dict(text='Tendencia Histórica: Gasto Total y % Variación Mes a Mes', x=0.5), # Título centrado
    yaxis=dict(title='Gasto Total (€)'),
    yaxis2=dict(
        title='Variación (%)',
        overlaying='y',
        side='right',
        showgrid=False,
        # Damos un poco de margen extra arriba para que el texto de la línea no se corte
        range=[df_macro['Variacion_Pct'].min() - 10, df_macro['Variacion_Pct'].max() + 20]
    ),

    # Leyenda abajo para no chocar con el título y completa
    legend=dict(
        orientation="h",
        yanchor="top",
        y=-0.15,               # Posición debajo del gráfico
        xanchor="center",
        x=0.5
    ),
    hovermode="x unified",
    # Ajustamos márgenes: para que quepa la leyenda en la parte inferior y título/texto en la superior
    margin=dict(t=80, b=100)
)

fig_macro.show()
```

```python
# @title Evolución del Gasto Total 2025 (Visión Acumulada)
import plotly.graph_objects as go
import pandas as pd

# 1. Preparación de Datos (Filtrado 2025 y Cálculo Acumulado)
# Aseguramos que trabajamos con una copia para no alterar el original
df_2025 = df[df['Date'].dt.year == 2025].copy()

# Agrupar por mes
df_evol_2025 = df_2025.groupby('YearMonth_Str')['Gasto_Abs'].sum().reset_index()
df_evol_2025 = df_evol_2025.sort_values('YearMonth_Str') # Asegurar orden cronológico

# CÁLCULO CLAVE: Gasto Acumulado (YTD - Year to Date)
df_evol_2025['Gasto_Acumulado'] = df_evol_2025['Gasto_Abs'].cumsum()

# 2. Construcción del Gráfico
fig_evol = go.Figure()

# Eje Y1 (Izquierdo): Barras de Gasto Mensual
fig_evol.add_trace(go.Bar(
    x=df_evol_2025['YearMonth_Str'],
    y=df_evol_2025['Gasto_Abs'],
    name='Gasto Mensual',
    marker_color='#003366', # NOMII Primary (Azul Profundo)
    text=df_evol_2025['Gasto_Abs'],
    texttemplate='%{y:,.0f}€',
    textposition='auto',
    # Estándar: Texto blanco y negrita para contraste en barras oscuras
    textfont=dict(color='white', weight='bold')
))

# Eje Y2 (Derecho): Línea de Gasto Acumulado
fig_evol.add_trace(go.Scatter(
    x=df_evol_2025['YearMonth_Str'],
    y=df_evol_2025['Gasto_Acumulado'],
    name='Gasto Acumulado (YTD)',
    yaxis='y2',
    mode='lines+markers+text',
    line=dict(color='#20C6B6', width=3), # NOMII Secondary (Turquesa) para crecimiento
    marker=dict(size=8, color='#20C6B6'),
    text=df_evol_2025['Gasto_Acumulado'],
    texttemplate='%{y:.2s}', # Formato compacto (ej. 1.2M) para no saturar
    textposition="top center",
    # Estándar: Color de la marca para el texto de la línea
    textfont=dict(color='#FFCC00', size=11),
    cliponaxis=False
))

# 3. Layout (Diseño Estandarizado)
fig_evol.update_layout(
    title=dict(
        text='Evolución del Gasto Total - Año 2025 (Visión anual acumulada)',
        x=0.5,
        font=dict(size=20, color='#003366')
    ),
    yaxis=dict(
        title='Gasto Mensual (€)',
        titlefont=dict(color='#003366'),
        tickfont=dict(color='#333333')
    ),
    yaxis2=dict(
        title='Acumulado Anual (€)',
        titlefont=dict(color='#003366'),
        tickfont=dict(color='#20C6B6'),
        overlaying='y',
        side='right',
        showgrid=False # Limpieza visual
    ),
    # Leyenda con espaciado extra solicitado
    legend=dict(
        orientation="h",
        yanchor="top",
        y=-0.2,
        xanchor="center",
        x=0.5,

    ),
    hovermode="x unified",
    margin=dict(t=100, b=60) # Margen inferior amplio para la leyenda espaciada
)

fig_evol.show()
```

```python
# @title Estructura de Rigidez de Costos: Fijo vs. Variable vs. Único (Evolución Mensual)
import plotly.graph_objects as go
import pandas as pd

# 1. Preparación de Datos
df_struc = df.copy()

# Filtramos solo 2025
df_struc = df_struc[df_struc['Date'].dt.year == 2025]
df_struc['YearMonth_Str'] = df_struc['Date'].dt.strftime('%Y-%m')

# Agrupar por Mes y Comportamiento de Costo
df_grouped = df_struc.groupby(['YearMonth_Str', 'Cost Behavior'])['Gasto_Abs'].sum().reset_index()

# Pivotear para tener columnas por tipo de costo
df_pivot = df_grouped.pivot(index='YearMonth_Str', columns='Cost Behavior', values='Gasto_Abs').fillna(0).reset_index()

# Calcular porcentajes para visualización 100% Stacked (Opcional, aquí usaremos absoluto para ver volumen real)
# Si prefieres ver % relativo, avísame para ajustar.

# 2. Construcción del Gráfico
fig_cost = go.Figure()

# Definición de colores semánticos
colors = {
    'Costo fijo': '#003366',      # Azul oscuro (Rigidez, base estructural)
    'Costo variable': '#20C6B6',  # Turquesa (Flexibilidad)
    'Costo único': '#FFCC00'      # Amarillo (Eventualidad, alerta)
}

# Iteramos por las categorías presentes para asegurar orden lógico
orden_capas = ['Costo fijo', 'Costo variable', 'Costo único'] # Fijo abajo, Variable medio, Único arriba

for tipo in orden_capas:
    if tipo in df_pivot.columns:
        fig_cost.add_trace(go.Scatter(
            x=df_pivot['YearMonth_Str'],
            y=df_pivot[tipo],
            name=tipo,
            mode='lines',
            stackgroup='one', # Esto crea el gráfico de área apilada
            line=dict(width=0.5),
            marker=dict(size=4),
            fillcolor=colors.get(tipo, '#999999'), # Color por defecto si falla el diccionario
            line_color=colors.get(tipo, '#333333')
        ))

# 3. Layout
fig_cost.update_layout(
    title=dict(
        text='Evolución de la Estructura de Costos (Rigidez Operativa)',
        x=0.5,
        font=dict(size=20, color='#003366')
    ),
    yaxis=dict(
        title='Gasto Mensual (€)',
        titlefont=dict(color='#003366'),
        tickfont=dict(color='#333333')
    ),
    legend=dict(
        orientation="h",
        yanchor="top",
        y=-0.15,
        xanchor="center",
        x=0.5
    ),
    hovermode="x unified",
    margin=dict(t=80, b=80),
    plot_bgcolor='#F9F9F9'
)

fig_cost.show()
```

```python
# @title Evolución del Gasto por Cuatrimestre (Visión Acumulada)
import plotly.graph_objects as go
import pandas as pd

# 1. Función para ordenar cronológicamente los cuatrimestres (Necesaria para que el acumulado tenga sentido)
def obtener_orden_cuatrimestre(val_cuatrimestre):
    try:
        # Asume formato "C1-2025" -> split por '-'
        partes = str(val_cuatrimestre).split('-')
        # Limpia prefijos para obtener el número del cuatrimestre
        prefijo = partes[0].upper().replace('C', '').replace('Q', '').replace('T', '')
        c = int(prefijo)
        y = int(partes[1])
        # Retorna valor entero para ordenar: 20251, 20252, etc.
        return y * 10 + c
    except:
        return 0

# 2. Preparación de Datos
# Usamos df.copy() como solicitado
df_evol_cuatri = df.copy()

# Agrupar por Cuatrimestre
df_evol_cuatri = df_evol_cuatri.groupby('Cuatrimestre')['Gasto_Abs'].sum().reset_index()

# Ordenar CRONOLÓGICAMENTE (Crucial para el cálculo acumulado)
df_evol_cuatri['Sort_Key'] = df_evol_cuatri['Cuatrimestre'].apply(obtener_orden_cuatrimestre)
df_evol_cuatri = df_evol_cuatri.sort_values('Sort_Key')

# CÁLCULO CLAVE: Gasto Acumulado
df_evol_cuatri['Gasto_Acumulado'] = df_evol_cuatri['Gasto_Abs'].cumsum()

# 3. Construcción del Gráfico
fig_evol = go.Figure()

# Eje Y1 (Izquierdo): Barras de Gasto por Cuatrimestre
fig_evol.add_trace(go.Bar(
    x=df_evol_cuatri['Cuatrimestre'],
    y=df_evol_cuatri['Gasto_Abs'],
    name='Gasto del Periodo',
    marker_color='#003366', # NOMII Primary (Azul Profundo)
    text=df_evol_cuatri['Gasto_Abs'],
    texttemplate='%{y:,.0f}€',
    textposition='auto',
    # Estándar: Texto blanco y negrita para contraste en barras oscuras
    textfont=dict(color='white', weight='bold')
))

# Eje Y2 (Derecho): Línea de Gasto Acumulado
fig_evol.add_trace(go.Scatter(
    x=df_evol_cuatri['Cuatrimestre'],
    y=df_evol_cuatri['Gasto_Acumulado'],
    name='Gasto Acumulado',
    yaxis='y2',
    mode='lines+markers+text',
    line=dict(color='#20C6B6', width=3), # NOMII Secondary (Turquesa)
    marker=dict(size=8, color='#20C6B6'),
    text=df_evol_cuatri['Gasto_Acumulado'],
    texttemplate='%{y:.2s}', # Formato compacto (ej. 1.2M)
    textposition="top center",
    # Estándar: Color de contraste (Amarillo/Dorado) para destacar el acumulado
    textfont=dict(color='#FFCC00', size=11, weight='bold'),
    cliponaxis=False
))

# 4. Layout (Diseño Estandarizado)
fig_evol.update_layout(
    title=dict(
        text='Evolución del Gasto por Cuatrimestre (Visión Acumulada)',
        x=0.5,
        font=dict(size=20, color='#003366')
    ),
    yaxis=dict(
        title='Gasto del Periodo (€)',
        titlefont=dict(color='#003366'),
        tickfont=dict(color='#333333')
    ),
    yaxis2=dict(
        title='Acumulado Total (€)',
        titlefont=dict(color='#003366'),
        tickfont=dict(color='#20C6B6'),
        overlaying='y',
        side='right',
        showgrid=False # Limpieza visual
    ),
    legend=dict(
        orientation="h",
        yanchor="top",
        y=-0.2,
        xanchor="center",
        x=0.5,
    ),
    hovermode="x unified",
    margin=dict(t=100, b=80) # Margen inferior amplio para la leyenda
)

fig_evol.show()
```

```python
# @title Proyección 2026: Holt-Winters (Tendencia + Estacionalidad Automática)
import plotly.graph_objects as go
import pandas as pd
from statsmodels.tsa.holtwinters import ExponentialSmoothing

# 1. Preparación de Datos (Serie de Tiempo Estricta)
df_hw = df.copy()

# Agrupar y asegurar índice de fecha con frecuencia mensual ('MS' = Month Start)
df_serie = df_hw.groupby(pd.Grouper(key='Date', freq='MS'))['Gasto_Abs'].sum()

# Rellenar meses vacíos con 0 o interpolación para no romper el modelo
df_serie = df_serie.fillna(0)

# Verificación de datos suficientes (Holt-Winters necesita al menos 2 ciclos/24 meses idealmente)
# Si hay pocos datos, usamos un fallback simple, si hay bastantes, usamos HW completo.
usar_estacionalidad = True if len(df_serie) >= 24 else False

# 2. Ajuste del Modelo y Predicción
try:
    if usar_estacionalidad:
        # Modelo Triple: Trend (Tendencia) + Seasonal (Estacionalidad)
        modelo = ExponentialSmoothing(
            df_serie,
            trend='add',
            seasonal='add',
            seasonal_periods=12
        ).fit()
        titulo_modelo = "Holt-Winters (Triple Suavizado)"
    else:
        # Si hay pocos datos, usamos solo Tendencia (Holt Lineal)
        modelo = ExponentialSmoothing(
            df_serie,
            trend='add',
            seasonal=None
        ).fit()
        titulo_modelo = "Holt Lineal (Datos históricos insuficientes para estacionalidad)"

    # Predecir los 12 meses de 2026
    # Calculamos cuántos pasos adelante necesitamos.
    # Si tus datos terminan en Dic 2025, forecast(12) predice todo 2026.
    prediccion_2026 = modelo.forecast(12)

except Exception as e:
    print(f"Error al ajustar modelo: {e}")
    # Fallback de seguridad: Promedio simple
    prediccion_2026 = pd.Series([df_serie.mean()] * 12, index=pd.date_range('2026-01-01', periods=12, freq='MS'))
    titulo_modelo = "Promedio Simple (Fallback por error en modelo)"

# 3. Estructurar Datos para Plotly
df_proyeccion = pd.DataFrame({
    'Fecha': prediccion_2026.index,
    'YearMonth_Str': prediccion_2026.index.strftime('%Y-%m'),
    'Gasto_Proyectado': prediccion_2026.values
})

# Evitar valores negativos (los modelos estadísticos a veces pueden darlos si la tendencia es muy a la baja)
df_proyeccion['Gasto_Proyectado'] = df_proyeccion['Gasto_Proyectado'].apply(lambda x: max(0, x))

# Cálculo Acumulado
df_proyeccion['Acumulado_Proyectado'] = df_proyeccion['Gasto_Proyectado'].cumsum()

# 4. Construcción del Gráfico
fig_hw = go.Figure()

# Eje Y1 (Izquierdo): Barras de Gasto Mensual
fig_hw.add_trace(go.Bar(
    x=df_proyeccion['YearMonth_Str'],
    y=df_proyeccion['Gasto_Proyectado'],
    name='Gasto Proyectado (HW)',
    marker_color='#003366', # NOMII Primary
    text=df_proyeccion['Gasto_Proyectado'],
    texttemplate='%{y:,.0f}€',
    textposition='auto',
    textfont=dict(color='white', weight='bold')
))

# Eje Y2 (Derecho): Línea de Acumulado
fig_hw.add_trace(go.Scatter(
    x=df_proyeccion['YearMonth_Str'],
    y=df_proyeccion['Acumulado_Proyectado'],
    name='Acumulado Anual',
    yaxis='y2',
    mode='lines+markers+text',
    line=dict(color='#20C6B6', width=3, dash='dot'),
    marker=dict(size=8, color='#20C6B6'),
    text=df_proyeccion['Acumulado_Proyectado'],
    texttemplate='%{y:.2s}',
    textposition="top center",
    textfont=dict(color='#FFCC00', size=11, weight='bold'),
    cliponaxis=False
))

# 5. Layout (Estilo Nomii)
fig_hw.update_layout(
    title=dict(
        text=f'Proyección 2026: {titulo_modelo}',
        x=0.5,
        font=dict(size=20, color='#003366')
    ),
    yaxis=dict(
        title='Gasto Mensual (€)',
        titlefont=dict(color='#003366'),
        tickfont=dict(color='#333333')
    ),
    yaxis2=dict(
        title='Acumulado Proyectado (€)',
        titlefont=dict(color='#003366'),
        tickfont=dict(color='#20C6B6'),
        overlaying='y',
        side='right',
        showgrid=False
    ),
    legend=dict(
        orientation="h",
        yanchor="top",
        y=-0.2,
        xanchor="center",
        x=0.5,
    ),
    hovermode="x unified",
    margin=dict(t=100, b=80)
)

fig_hw.show()
```

```python
# @title Proyección 2026: Holt-Winters (Tendencia + Estacionalidad Automática) (Filtrando costos únicos)
import plotly.graph_objects as go
import pandas as pd
from statsmodels.tsa.holtwinters import ExponentialSmoothing

# 1. Preparación de Datos (Serie de Tiempo Estricta)
df_hw = df.copy()

# APLICACIÓN DEL FILTRO: Excluir registros marcados como 'Costo único'
# Fundamental para que los "picos" de gastos únicos no confundan al modelo estadístico
df_hw = df_hw[df_hw['Cost Behavior'] != 'Costo único']

# Agrupar y asegurar índice de fecha con frecuencia mensual ('MS' = Month Start)
df_serie = df_hw.groupby(pd.Grouper(key='Date', freq='MS'))['Gasto_Abs'].sum()

# Rellenar meses vacíos con 0 o interpolación para no romper el modelo
df_serie = df_serie.fillna(0)

# Verificación de datos suficientes (Holt-Winters necesita al menos 2 ciclos/24 meses idealmente)
# Si hay pocos datos, usamos un fallback simple, si hay bastantes, usamos HW completo.
usar_estacionalidad = True if len(df_serie) >= 24 else False

# 2. Ajuste del Modelo y Predicción
try:
    if usar_estacionalidad:
        # Modelo Triple: Trend (Tendencia) + Seasonal (Estacionalidad)
        modelo = ExponentialSmoothing(
            df_serie,
            trend='add',
            seasonal='add',
            seasonal_periods=12
        ).fit()
        titulo_modelo = "Holt-Winters (Triple Suavizado)"
    else:
        # Si hay pocos datos, usamos solo Tendencia (Holt Lineal)
        modelo = ExponentialSmoothing(
            df_serie,
            trend='add',
            seasonal=None
        ).fit()
        titulo_modelo = "Holt Lineal (Datos históricos insuficientes para estacionalidad) (Filtrando costos únicos)"

    # Predecir los 12 meses de 2026
    # Calculamos cuántos pasos adelante necesitamos.
    # Si tus datos terminan en Dic 2025, forecast(12) predice todo 2026.
    prediccion_2026 = modelo.forecast(12)

except Exception as e:
    print(f"Error al ajustar modelo: {e}")
    # Fallback de seguridad: Promedio simple
    prediccion_2026 = pd.Series([df_serie.mean()] * 12, index=pd.date_range('2026-01-01', periods=12, freq='MS'))
    titulo_modelo = "Promedio Simple (Fallback por error en modelo)"

# 3. Estructurar Datos para Plotly
df_proyeccion = pd.DataFrame({
    'Fecha': prediccion_2026.index,
    'YearMonth_Str': prediccion_2026.index.strftime('%Y-%m'),
    'Gasto_Proyectado': prediccion_2026.values
})

# Evitar valores negativos (los modelos estadísticos a veces pueden darlos si la tendencia es muy a la baja)
df_proyeccion['Gasto_Proyectado'] = df_proyeccion['Gasto_Proyectado'].apply(lambda x: max(0, x))

# Cálculo Acumulado
df_proyeccion['Acumulado_Proyectado'] = df_proyeccion['Gasto_Proyectado'].cumsum()

# 4. Construcción del Gráfico
fig_hw = go.Figure()

# Eje Y1 (Izquierdo): Barras de Gasto Mensual
fig_hw.add_trace(go.Bar(
    x=df_proyeccion['YearMonth_Str'],
    y=df_proyeccion['Gasto_Proyectado'],
    name='Gasto Proyectado (HW)',
    marker_color='#003366', # NOMII Primary
    text=df_proyeccion['Gasto_Proyectado'],
    texttemplate='%{y:,.0f}€',
    textposition='auto',
    textfont=dict(color='white', weight='bold')
))

# Eje Y2 (Derecho): Línea de Acumulado
fig_hw.add_trace(go.Scatter(
    x=df_proyeccion['YearMonth_Str'],
    y=df_proyeccion['Acumulado_Proyectado'],
    name='Acumulado Anual',
    yaxis='y2',
    mode='lines+markers+text',
    line=dict(color='#20C6B6', width=3, dash='dot'),
    marker=dict(size=8, color='#20C6B6'),
    text=df_proyeccion['Acumulado_Proyectado'],
    texttemplate='%{y:.2s}',
    textposition="top center",
    textfont=dict(color='#FFCC00', size=11, weight='bold'),
    cliponaxis=False
))

# 5. Layout (Estilo Nomii)
fig_hw.update_layout(
    title=dict(
        text=f'Proyección 2026: {titulo_modelo}',
        x=0.5,
        font=dict(size=20, color='#003366')
    ),
    yaxis=dict(
        title='Gasto Mensual (€)',
        titlefont=dict(color='#003366'),
        tickfont=dict(color='#333333')
    ),
    yaxis2=dict(
        title='Acumulado Proyectado (€)',
        titlefont=dict(color='#003366'),
        tickfont=dict(color='#20C6B6'),
        overlaying='y',
        side='right',
        showgrid=False
    ),
    legend=dict(
        orientation="h",
        yanchor="top",
        y=-0.2,
        xanchor="center",
        x=0.5,
    ),
    hovermode="x unified",
    margin=dict(t=100, b=80)
)

fig_hw.show()
```

```python
# @title Comparativa: Gasto por Departamento entre Periodos (Cuatrimestre)
import plotly.graph_objects as go
import plotly.express as px

# 1. Configuración de Estilo y Colores Nomii
colores_periodos = [
    '#003366', # C1: Azul Profundo (Principal)
    '#20C6B6', # C2: Turquesa (Destacado)
    '#2E86C1', # C3: Azul Intermedio
    '#1ABC9C'  # C4: Turquesa Oscuro
]

# 2. Función de Ordenamiento (Reutilizada para consistencia)
def obtener_orden_cuatrimestre(val_cuatrimestre):
    try:
        partes = str(val_cuatrimestre).split('-')
        prefijo = partes[0].upper().replace('C', '').replace('Q', '').replace('T', '')
        c = int(prefijo)
        y = int(partes[1])
        return y * 10 + c
    except:
        return 0

# 3. Preparación de Datos
df_comp = df.copy()

# Verificar que existe la columna (fallback a trimestres si no existe cuatrimestre)
columna_periodo = 'Cuatrimestre' if 'Cuatrimestre' in df_comp.columns else 'Period'

if columna_periodo in df_comp.columns:
    # Agrupar por Departamento y Periodo
    df_agrupado = df_comp.groupby(['Department', columna_periodo])['Gasto_Abs'].sum().reset_index()

    # Ordenar los periodos cronológicamente
    df_agrupado['Sort_Key'] = df_agrupado[columna_periodo].apply(obtener_orden_cuatrimestre)
    df_agrupado = df_agrupado.sort_values(['Sort_Key'])

    # Calcular el gasto total por departamento para ordenar el Eje X (De mayor a menor gasto)
    top_deptos = df_comp.groupby('Department')['Gasto_Abs'].sum().sort_values(ascending=False).index.tolist()

    # 4. Construcción del Gráfico
    fig_comp = go.Figure()

    # Define NOMII blue color locally for text differentiation
    nomii_blue = '#003366'

    # Obtener lista única de periodos ordenados para iterar
    periodos_unicos = df_agrupado[columna_periodo].unique()

    # Crear una barra para cada periodo
    for i, periodo in enumerate(periodos_unicos):
        # Filtrar datos de ese periodo
        data_periodo = df_agrupado[df_agrupado[columna_periodo] == periodo]

        # Asignar color cíclico de la paleta
        color_trace = colores_periodos[i % len(colores_periodos)]

        # Determine text colors based on department
        text_colors_for_trace = []
        for dept_name in data_periodo['Department']:
            if dept_name in ['Gerencia', 'Servicio al Cliente', 'Design']:
                text_colors_for_trace.append(nomii_blue)
            else:
                text_colors_for_trace.append('white')

        fig_comp.add_trace(go.Bar(
            x=data_periodo['Department'],
            y=data_periodo['Gasto_Abs'],
            name=str(periodo),
            marker_color=color_trace,
            text=data_periodo['Gasto_Abs'],
            texttemplate='%{y:,.0f}€',
            textposition='auto',
            textfont=dict(color=text_colors_for_trace, size=10)
        ))

    # 5. Layout (Estilo Nomii)
    fig_comp.update_layout(
        title=dict(
            text='Comparativa de Gasto por Departamento y Periodo',
            x=0.5,
            font=dict(size=20, color='#003366')
        ),
        xaxis=dict(
            title='Departamento',
            categoryorder='array', # Forzar orden específico
            categoryarray=top_deptos, # Ordenar departamentos de mayor a menor gasto total
            tickangle=-45 # Inclinar etiquetas si son largas
        ),
        yaxis=dict(
            title='Monto (€)',
            titlefont=dict(color='#003366'),
            tickfont=dict(color='#333333'),
            gridcolor='#e0e0e0'
        ),
        barmode='group', # CLAVE: Barras agrupadas lado a lado
        legend=dict(
            orientation="h",
            yanchor="bottom",
            y=1.02,
            xanchor="right",
            x=1,
            title="Periodo"
        ),
        hovermode="x unified",
        margin=dict(t=80, b=100) # Margen inferior extra para nombres de deptos
    )

    fig_comp.show()

else:
    print(f"Error: No se encontró la columna '{columna_periodo}' para realizar la comparación.")
```

```python
# @title Composición del Gasto 2025 (Mix de Categorías - Donut Chart)
import plotly.graph_objects as go
import pandas as pd

# 1. Preparación de Datos
df_mix = df.copy()

# Filtrar solo el año 2025
df_mix = df_mix[df_mix['Date'].dt.year == 2025]

# Definir columna de categoría (Ajusta 'Category' si tu columna tiene otro nombre)
col_categoria = 'Category'

if col_categoria in df_mix.columns:
    # Agrupar por categoría
    df_grouped = df_mix.groupby(col_categoria)['Gasto_Abs'].sum().reset_index()

    # Calcular el total para sacar porcentajes
    total_gasto = df_grouped['Gasto_Abs'].sum()

    # --- LÓGICA DE AGRUPACIÓN "OTROS" ---
    # Si una categoría es menor al 3% del total, se mueve a "Otros"
    umbral_corte = 0.03 # 3%

    df_grouped['Porcentaje'] = df_grouped['Gasto_Abs'] / total_gasto

    # Separar principales y pequeños
    df_principales = df_grouped[df_grouped['Porcentaje'] >= umbral_corte].copy()
    df_otros = df_grouped[df_grouped['Porcentaje'] < umbral_corte].copy()

    # Si hay categorías pequeñas, crear la fila "Otros"
    if not df_otros.empty:
        fila_otros = pd.DataFrame({
            col_categoria: ['Otros'],
            'Gasto_Abs': [df_otros['Gasto_Abs'].sum()],
            'Porcentaje': [df_otros['Gasto_Abs'].sum() / total_gasto]
        })
        df_final = pd.concat([df_principales, fila_otros], ignore_index=True)
    else:
        df_final = df_principales

    # Ordenar de mayor a menor para que el gráfico empiece a las 12 en punto con el más grande
    df_final = df_final.sort_values('Gasto_Abs', ascending=False)

    # 2. Configuración de Colores Nomii
    # Paleta secuencial azul-turquesa para armonía
    colores_nomii_pie = [
        '#003366', # Azul Profundo (Categoría 1)
        '#20C6B6', # Turquesa (Categoría 2)
        '#2E86C1', # Azul Medio
        '#1ABC9C', # Turquesa Oscuro
        '#5D6D7E', # Gris Azulado
        '#AED6F1', # Azul Claro
        '#A3E4D7', # Turquesa Claro
        '#BDC3C7'  # Gris (Usualmente para "Otros")
    ]

    # 3. Construcción del Gráfico
    fig_pie = go.Figure()

    fig_pie.add_trace(go.Pie(
        labels=df_final[col_categoria],
        values=df_final['Gasto_Abs'],
        hole=0.45, # Crea el efecto Donut
        marker=dict(colors=colores_nomii_pie),
        textinfo='percent+label', # Muestra etiqueta y %
        textposition='outside',   # Etiquetas fuera para mejor lectura
        hoverinfo='label+value+percent',
        hovertemplate='<b>%{label}</b><br>Gasto: %{value:,.0f}€<br>Peso: %{percent}<extra></extra>'
    ))

    # 4. Layout (Estilo Nomii)
    fig_pie.update_layout(
        title=dict(
            text='Composición del Gasto 2025 (Mix de Categorías)',
            x=0.5,
            font=dict(size=20, color='#003366')
        ),
        # Añadir el Total en el centro del Donut
        annotations=[dict(
            text=f'Total 2025<br><b>{total_gasto:,.0f}€</b>', # Mostrar el total sin truncar a M€
            x=0.5, y=0.5,
            font_size=16,
            font_color='#003366',
            showarrow=False
        )],
        legend=dict(
            orientation="h",
            yanchor="bottom",
            y=-0.15, # Leyenda abajo para no chocar con etiquetas laterales
            xanchor="center",
            x=0.5
        ),
        margin=dict(t=80, b=50, l=50, r=50)
    )

    fig_pie.show()

else:
    print(f"Error: No se encontró la columna '{col_categoria}'.")
```

```python
# @title Comparativa Estructural: Primer Mes vs Último Mes (Dumbbell Chart)
import plotly.graph_objects as go
import pandas as pd

# 1. Preparación de Datos
df_struct_mes = df.copy()

# Creamos una columna de Periodo para ordenar (YYYY-MM)
df_struct_mes['Periodo_Str'] = df_struct_mes['Date'].dt.strftime('%Y-%m')

# Identificar periodos disponibles ordenados
periodos_disponibles = sorted(df_struct_mes['Periodo_Str'].unique())

if len(periodos_disponibles) >= 2:
    # Seleccionamos el PRIMERO y el ÚLTIMO automáticamente
    periodo_inicial = periodos_disponibles[0]  # Ej: 2025-01
    periodo_final = periodos_disponibles[-1]   # Ej: 2025-11

    # Filtrar datos de estos dos periodos
    df_filtrado = df_struct_mes[df_struct_mes['Periodo_Str'].isin([periodo_inicial, periodo_final])]

    # Agrupar por Periodo y Categoría
    df_grouped = df_filtrado.groupby(['Periodo_Str', 'Category'])['Gasto_Abs'].sum().reset_index()
    df_grouped.columns = ['Periodo', 'Category', 'Gasto']

    # Calcular Totales por Periodo para sacar porcentajes (Share)
    total_por_periodo = df_grouped.groupby('Periodo')['Gasto'].transform('sum')
    df_grouped['Share'] = df_grouped['Gasto'] / total_por_periodo

    # Pivotar: [Category, Share_Inicial, Share_Final]
    df_pivot = df_grouped.pivot(index='Category', columns='Periodo', values='Share').fillna(0)

    # Asegurarnos de tener las columnas correctas (el pivot ordena alfabéticamente, pero queremos Inicio -> Fin)
    col_inicio = periodo_inicial
    col_fin = periodo_final

    # Calcular Delta (Cambio en Puntos Porcentuales)
    df_pivot['Delta'] = df_pivot[col_fin] - df_pivot[col_inicio]

    # Ordenar por el peso del mes actual (para visualizar mejor los gastos grandes hoy)
    df_pivot = df_pivot.sort_values(col_fin, ascending=True)

    categorias = df_pivot.index

    # 2. Construcción del Gráfico
    fig_dumb_mes = go.Figure()

    # A) Líneas conectores (Gris)
    for cat in categorias:
        row = df_pivot.loc[cat]
        fig_dumb_mes.add_trace(go.Scatter(
            x=[row[col_inicio], row[col_fin]],
            y=[cat, cat],
            mode='lines',
            line=dict(color='#BDC3C7', width=2),
            showlegend=False,
            hoverinfo='skip'
        ))

    # B) Punto Inicial (Gris - "Como empezamos")
    fig_dumb_mes.add_trace(go.Scatter(
        x=df_pivot[col_inicio],
        y=categorias,
        mode='markers',
        name=f"Inicio ({col_inicio})",
        marker=dict(color='#95A5A6', size=10, symbol='circle'),
        hovertemplate=f'<b>{col_inicio}</b>: %{{x:.1%}}<extra></extra>'
    ))

    # C) Punto Final (Azul Nomii - "Como estamos hoy")
    fig_dumb_mes.add_trace(go.Scatter(
        x=df_pivot[col_fin],
        y=categorias,
        mode='markers',
        name=f"Cierre ({col_fin})",
        marker=dict(color='#003366', size=14, symbol='circle'), # Azul Nomii
        hovertemplate=f'<b>{col_fin}</b>: %{{x:.1%}}<extra></extra>'
    ))

    # 3. Anotaciones de Cambio (Deltas)
    annotations = []
    for cat in categorias:
        row = df_pivot.loc[cat]
        # Determinar color y símbolo
        color_delta = '#20C6B6' if row['Delta'] >= 0 else '#E74C3C' # Turquesa vs Rojo
        simbolo = "+" if row['Delta'] >= 0 else ""

        # Solo poner etiqueta si el cambio es relevante (> 0.1%) para no ensuciar
        if abs(row['Delta']) > 0.001:
            annotations.append(dict(
                x=max(row[col_inicio], row[col_fin]),
                y=cat,
                text=f" <b>{simbolo}{row['Delta']*100:.1f} pp</b>",
                font=dict(color=color_delta, size=10),
                showarrow=False,
                xanchor='left',
                xshift=10
            ))

    fig_dumb_mes.update_layout(annotations=annotations)

    # 4. Layout (Estilo Nomii)
    fig_dumb_mes.update_layout(
        title=dict(
            text=f'Cambio en Estructura de Costos: {col_inicio} vs {col_fin}',
            x=0.5,
            font=dict(size=20, color='#003366')
        ),
        xaxis=dict(
            title='Peso sobre el Gasto Total (%)',
            tickformat='.0%',
            range=[0, df_pivot[[col_inicio, col_fin]].max().max() * 1.25], # Margen extra
            showgrid=True,
            gridcolor='#f0f0f0'
        ),
        yaxis=dict(
            title='',
            tickfont=dict(size=11)
        ),
        legend=dict(
            orientation="h",
            yanchor="bottom",
            y=1.02,
            xanchor="right",
            x=1
        ),
        margin=dict(t=100, r=50),
        height=max(500, len(categorias) * 40)
    )

    fig_dumb_mes.show()

else:
    print("⚠️ Error: Necesitas al menos 2 meses de datos para realizar esta comparativa.")
```

```python
# @title Proyección de Tesorería Controlada (Sin saltos)
from prophet import Prophet
import pandas as pd
import plotly.graph_objects as go
import numpy as np

# 1. Auditoría de Datos en Memoria (¿Seguimos con millones?)
# ---------------------------------------------------------
df_prophet = df.copy()

# A. Limpieza de Moneda (Redundancia de seguridad)
# Si encontramos algún valor > 100,000 (y es CLP), lo arreglamos al vuelo
def fix_currency_emergency(row):
    val = abs(row['Amount in EUR'])
    if val > 100000: # Si es mayor a 100k euros, sospechamos
        # Intentamos usar el Original Amount corregido
        return row['Original Amount'] * row['ER to EUR']
    if pd.isna(val):
        return row['Original Amount'] * row['ER to EUR']
    return row['Amount in EUR']

df_prophet['Amount in EUR'] = df_prophet.apply(fix_currency_emergency, axis=1)
df_prophet['Gasto_Abs'] = df_prophet['Amount in EUR'].abs()

# B. IMPRESIÓN DE DIAGNÓSTICO (¡Mira esto en la salida!)
top_gasto = df_prophet['Gasto_Abs'].max()
print(f"🔎 DIAGNÓSTICO: El gasto diario más alto en el historial es: €{top_gasto:,.2f}")
if top_gasto > 200000:
    print("⚠️ ALERTA: Aún hay datos de 'Millones' en tu memoria. El gráfico saldrá mal.")
else:
    print("✅ DATOS CORRECTOS: Los montos están en rango normal.")


# 2. Preparación para Prophet
# ---------------------------------------------------------
df_daily = df_prophet.groupby('Date')['Gasto_Abs'].sum().reset_index()
df_daily.columns = ['ds', 'y']

# TRUCO: Eliminamos outliers extremos (ej. gastos de >100k en un día) para no confundir al modelo
# df_daily = df_daily[df_daily['y'] < 100000]

# 3. Modelado con "Techo" (Growth = Logistic)
# ---------------------------------------------------------
# Esto evita que la proyección se vaya al infinito
df_daily['cap'] = 15000  # Decimos: "Es imposible gastar más de 15k PROMEDIO diario"
df_daily['floor'] = 0

m = Prophet(daily_seasonality=False,
            yearly_seasonality=True,
            weekly_seasonality=True,
            growth='logistic') # Cambiamos a logístico para frenar la explosión

m.fit(df_daily)

future = m.make_future_dataframe(periods=365)
future['cap'] = 15000 # Aplicamos el mismo techo al futuro
future['floor'] = 0

forecast = m.predict(future)

# 4. Visualización
# ---------------------------------------------------------
fig = go.Figure()

# Datos Reales
fig.add_trace(go.Scatter(
    x=df_daily['ds'], y=df_daily['y'],
    mode='markers', name='Gasto Real',
    marker=dict(color='#34495E', size=5, opacity=0.6)
))

# Proyección IA
fig.add_trace(go.Scatter(
    x=forecast['ds'], y=forecast['yhat'],
    mode='lines', name='Tendencia (Suavizada)',
    line=dict(color='#1ABC9C', width=3)
))

fig.update_layout(
    title="<b>Proyección de Flujo de Caja (Corregida)</b>",
    yaxis_title="Gasto Diario Estimado (€)",
    template="plotly_white",
    hovermode="x unified"
)

fig.show()
```

```python
# @title Mapa de Calor Mensual: Intensidad de Gasto (Día vs Mes)
import plotly.graph_objects as go
import pandas as pd

# --- CONFIGURACIÓN DE TAMAÑO ---
ancho_grafico = 1400
alto_grafico = 600

# 1. Preparación de Datos
df_cal = df.copy()

# Extraer componentes de fecha
df_cal['Mes_Num'] = df_cal['Date'].dt.month
df_cal['Mes_Nombre'] = df_cal['Date'].dt.strftime('%b') # Ej: Jan, Feb (o Ene, Feb según locale)
df_cal['Dia_Num'] = df_cal['Date'].dt.dayofweek # 0=Lunes, 6=Domingo
df_cal['Dia_Nombre'] = df_cal['Date'].dt.day_name() # Ej: Monday

# Agrupar datos: Suma de gasto por Mes y Día de la Semana
# Esto nos dice "Cuánto gastamos en total todos los Lunes de Enero"
cal_data = df_cal.groupby(['Mes_Num', 'Mes_Nombre', 'Dia_Num', 'Dia_Nombre'])['Gasto_Abs'].sum().reset_index()

# Pivotar para crear la matriz (Filas=Días, Columnas=Meses)
# Usamos pivot_table para manejar huecos con 0
matriz_cal = cal_data.pivot_table(
    index=['Dia_Num', 'Dia_Nombre'],
    columns=['Mes_Num', 'Mes_Nombre'],
    values='Gasto_Abs',
    fill_value=0
)

# 2. Extraer etiquetas ordenadas para los ejes
# Como el índice y columnas son MultiIndex (Num, Nombre), extraemos solo el nombre pero ordenado por el Num
dias_ordenados = matriz_cal.index.get_level_values(1).unique()
meses_ordenados = matriz_cal.columns.get_level_values(1).unique()

# Extraer solo los valores para el heatmap
z_values = matriz_cal.values

# 3. Construcción del Gráfico
fig_heat = go.Figure(data=go.Heatmap(
    z=z_values,
    x=meses_ordenados,
    y=dias_ordenados,
    colorscale=[
        [0, '#FFFFFF'],       # 0 = Blanco (Sin actividad)
        [0.2, '#D1F2EB'],     # Bajo = Turquesa muy pálido
        [0.5, '#20C6B6'],     # Medio = Turquesa Nomii
        [1, '#003366']        # Alto = Azul Profundo
    ],
    xgap=2, # Espacio entre celdas horiz
    ygap=2, # Espacio entre celdas vert
    hovertemplate='<b>%{y}, %{x}</b><br>Gasto Total: %{z:,.0f}€<extra></extra>'
))

# 4. Layout (Estilo Nomii)
fig_heat.update_layout(
    title=dict(
        text='Intensidad de Gasto: Días de la Semana vs. Meses',
        x=0.5,
        font=dict(size=20, color='#003366')
    ),
    width=ancho_grafico,
    height=alto_grafico,
    xaxis=dict(
        title='Mes del Año',
        titlefont=dict(color='#003366'),
        side='bottom'
    ),
    yaxis=dict(
        title='',
        tickfont=dict(color='#333333'),
        # IMPORTANTE: Invertimos el eje Y para que Lunes quede arriba (estilo calendario)
        autorange='reversed'
    ),
    plot_bgcolor='white'
)

fig_heat.show()
```

```python
# @title Auditoría Automática: Detección de Anomalías (Isolation Forest)
import pandas as pd
import plotly.graph_objects as go
from sklearn.ensemble import IsolationForest

# --- CONFIGURACIÓN DE COLORES INSTITUCIONALES ---
COLOR_NORMAL = "#2C3E50"    # Azul Oscuro / Gris (Transacciones normales)
COLOR_ALERTA = "#E74C3C"    # Rojo (Anomalías detectadas)
COLOR_FONDO  = "white"      # Fondo del gráfico

# 1. Preparación de Datos
# ---------------------------------------------------------
df_audit = df.copy()

# Limpieza y Feature Engineering
df_audit['Amount_Clean'] = df_audit['Amount in EUR'].fillna(0)
df_audit['Dia_Semana'] = df_audit['Date'].dt.dayofweek
df_audit['Dia_Anio'] = df_audit['Date'].dt.dayofyear

# Seleccionar variables para el modelo (Monto y Patrones de Fecha)
features = df_audit[['Amount_Clean', 'Dia_Semana', 'Dia_Anio']]

# 2. Entrenamiento del Modelo (Isolation Forest)
# ---------------------------------------------------------
# contamination=0.03 -> Asume que aprox el 3% de los datos podrían ser errores/fraude
iso_forest = IsolationForest(contamination=0.03, random_state=42)
df_audit['Outlier_Status'] = iso_forest.fit_predict(features)

# Separar data para graficar
df_normal = df_audit[df_audit['Outlier_Status'] == 1]
df_anomalia = df_audit[df_audit['Outlier_Status'] == -1]

# 3. Visualización Corporativa
# ---------------------------------------------------------
fig = go.Figure()

# A. Puntos Normales
fig.add_trace(go.Scatter(
    x=df_normal['Date'],
    y=df_normal['Amount in EUR'],
    mode='markers',
    name='Transacción Normal',
    marker=dict(color=COLOR_NORMAL, opacity=0.6, size=6),
    text=df_normal['Description'],
    hovertemplate="<b>%{text}</b><br>Fecha: %{x}<br>Monto: €%{y:.2f}"
))

# B. Anomalías (Alertas)
fig.add_trace(go.Scatter(
    x=df_anomalia['Date'],
    y=df_anomalia['Amount in EUR'],
    mode='markers',
    name='Posible Anomalía',
    marker=dict(color=COLOR_ALERTA, opacity=0.9, size=10, line=dict(width=2, color='white')),
    text=df_anomalia['Description'] + " - " + df_anomalia['Counterparty'],
    hovertemplate="<b>⚠️ ALERTA: %{text}</b><br>Fecha: %{x}<br>Monto: €%{y:.2f}"
))

# C. Layout Institucional
fig.update_layout(
    title="<b>Auditoría Inteligente:</b> Detección de Gastos Atípicos",
    xaxis_title="Fecha de Transacción",
    yaxis_title="Monto (€)",
    template="plotly_white",
    hovermode="closest",
    legend=dict(orientation="h", yanchor="bottom", y=1.02, xanchor="right", x=1)
)

fig.show()

# 4. Tabla Resumen de Alertas (Top 10)
print("\n--- ⚠️ TOP 10 TRANSACCIONES PARA REVISIÓN MANUAL ---")
cols_mostrar = ['Date', 'Counterparty', 'Description', 'Amount in EUR', 'Category']
display(df_anomalia.sort_values(by='Amount in EUR', ascending=True)[cols_mostrar].head(10))
```
